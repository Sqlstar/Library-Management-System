```python
%%writefile app.py
import streamlit as st
import cx_Oracle

import streamlit as st
import cx_Oracle

# Function to connect to Oracle Database
def connect_to_db():
    connection = cx_Oracle.connect(user="demouser", password="sukarna", dsn="DESKTOP-HPBF0ID:1521/XEPDB1")
    return connection

# Function to insert a publisher
def insert_publisher(publisher_name, publisher_address, publisher_phone):
    conn = connect_to_db()
    cursor = conn.cursor()
    try:
        cursor.execute("""
            INSERT INTO table_publisher (publisher_name, publisher_address, publisher_phone)
            VALUES (:1, :2, :3)
        """, (publisher_name, publisher_address, publisher_phone))
        conn.commit()
        st.success("Publisher inserted successfully!")
    except cx_Oracle.DatabaseError as e:
        st.error(f"Error: {e}")
    finally:
        cursor.close()
        conn.close()

# Function to insert a book
def insert_book(book_title, publisher_name):
    conn = connect_to_db()
    cursor = conn.cursor()
    try:
        cursor.execute("""
            INSERT INTO table_book (BookID, Book_Title, PublisherName)
            VALUES (book_id_seq.NEXTVAL, :1, :2)
        """, (book_title, publisher_name))
        conn.commit()
        st.success("Book inserted successfully!")
    except cx_Oracle.DatabaseError as e:
        st.error(f"Error: {e}")
    finally:
        cursor.close()
        conn.close()

# Function to insert a library branch
def insert_library_branch(branch_name, branch_address):
    conn = connect_to_db()
    cursor = conn.cursor()
    try:
        cursor.execute("""
            INSERT INTO table_library_branch (library_branch_BranchID, library_branch_BranchName, library_branch_BranchAddress)
            VALUES (branch_id_seq.NEXTVAL, :1, :2)
        """, (branch_name, branch_address))
        conn.commit()
        st.success("Library branch inserted successfully!")
    except cx_Oracle.DatabaseError as e:
        st.error(f"Error: {e}")
    finally:
        cursor.close()
        conn.close()

# Function to insert a borrower
def insert_borrower(borrower_name, borrower_address, borrower_phone):
    conn = connect_to_db()
    cursor = conn.cursor()
    try:
        cursor.execute("""
            INSERT INTO table_borrower (CardNo, BorrowerName, BorrowerAddress, BorrowerPhone)
            VALUES (borrower_seq.NEXTVAL, :1, :2, :3)
        """, (borrower_name, borrower_address, borrower_phone))
        conn.commit()
        st.success("Borrower inserted successfully!")
    except cx_Oracle.DatabaseError as e:
        st.error(f"Error: {e}")
    finally:
        cursor.close()
        conn.close()

# Function to insert book copies
def insert_book_copies(book_id, branch_id, no_of_copies):
    conn = connect_to_db()
    cursor = conn.cursor()
    try:
        cursor.execute("""
            INSERT INTO table_book_copies (CopiesID, BookID, BranchID, No_Of_Copies)
            VALUES (copies_seq.NEXTVAL, :1, :2, :3)
        """, (book_id, branch_id, no_of_copies))
        conn.commit()
        st.success("Book copies inserted successfully!")
    except cx_Oracle.DatabaseError as e:
        st.error(f"Error: {e}")
    finally:
        cursor.close()
        conn.close()

# Function to insert book authors
def insert_book_authors(book_id, author_name):
    conn = connect_to_db()
    cursor = conn.cursor()
    try:
        cursor.execute("""
            INSERT INTO table_book_authors (AuthorID, BookID, AuthorName)
            VALUES (author_seq.NEXTVAL, :1, :2)
        """, (book_id, author_name))
        conn.commit()
        st.success("Book author inserted successfully!")
    except cx_Oracle.DatabaseError as e:
        st.error(f"Error: {e}")
    finally:
        cursor.close()
        conn.close()

# Function to insert book details
def insert_book_details(isbn, title, author, category, price, copies):
    conn = connect_to_db()
    cursor = conn.cursor()
    try:
        cursor.execute("""
            INSERT INTO book (isbn, title, author, category, price, copies)
            VALUES (:1, :2, :3, :4, :5, :6)
        """, (isbn, title, author, category, price, copies))
        conn.commit()
        st.success("Book details inserted successfully!")
    except cx_Oracle.DatabaseError as e:
        st.error(f"Error: {e}")
    finally:
        cursor.close()
        conn.close()

# Function to insert book issue
def insert_book_issue(member, book_isbn, price):
    conn = connect_to_db()
    cursor = conn.cursor()
    try:
        cursor.execute("""
            INSERT INTO book_issue (issue_id, member, book_isbn, due_date, price)
            VALUES (book_issue_seq.NEXTVAL, :1, :2, SYSDATE + 20, :3)
        """, (member, book_isbn, price))
        conn.commit()
        st.success("Book issue recorded successfully!")
    except cx_Oracle.DatabaseError as e:
        st.error(f"Error: {e}")
    finally:
        cursor.close()
        conn.close()

# Function to insert librarian
def insert_librarian(username, password):
    conn = connect_to_db()
    cursor = conn.cursor()
    try:
        cursor.execute("""
            INSERT INTO librarian (id, username, password)
            VALUES (librarian_seq.NEXTVAL, :1, :2)
        """, (username, password))
        conn.commit()
        st.success("Librarian inserted successfully!")
    except cx_Oracle.DatabaseError as e:
        st.error(f"Error: {e}")
    finally:
        cursor.close()
        conn.close()

# Function to insert pending book requests
def insert_pending_book_request(member, book_isbn):
    conn = connect_to_db()
    cursor = conn.cursor()
    try:
        cursor.execute("""
            INSERT INTO pending_book_requests (request_id, member, book_isbn, time)
            VALUES (pending_request_seq.NEXTVAL, :1, :2, CURRENT_TIMESTAMP)
        """, (member, book_isbn))
        conn.commit()
        st.success("Pending book request added successfully!")
    except cx_Oracle.DatabaseError as e:
        st.error(f"Error: {e}")
    finally:
        cursor.close()
        conn.close()

# Function to insert pending registration
def insert_pending_registration(username, password, name, email, balance):
    conn = connect_to_db()
    cursor = conn.cursor()
    try:
        cursor.execute("""
            INSERT INTO pending_registrations (username, password, name, email, balance, time)
            VALUES (:1, :2, :3, :4, :5, CURRENT_TIMESTAMP)
        """, (username, password, name, email, balance))
        conn.commit()
        st.success("Pending registration added successfully!")
    except cx_Oracle.DatabaseError as e:
        st.error(f"Error: {e}")
    finally:
        cursor.close()
        conn.close()

# Streamlit app layout
def main():
    st.title("Library Management System")

    # Tabs for various operations
    tab1, tab2, tab3, tab4, tab5, tab6, tab7, tab8, tab9 = st.tabs([
        "Add Publisher", "Add Book", "Add Library Branch", "Add Borrower", 
        "Add Book Copies", "Add Book Authors", "Issue Book", "Add Librarian", "Pending Registration"
    ])

    # Tab 1: Add Publisher
    with tab1:
        st.header("Insert a New Publisher")
        publisher_name = st.text_input("Publisher Name", key="pub_name_1")
        publisher_address = st.text_input("Publisher Address", key="pub_address_1")
        publisher_phone = st.text_input("Publisher Phone", key="pub_phone_1")
        if st.button("Add Publisher", key="add_pub_btn"):
            insert_publisher(publisher_name, publisher_address, publisher_phone)

    # Tab 2: Add Book
    with tab2:
        st.header("Insert a New Book")
        book_title = st.text_input("Book Title", key="book_title_2")
        publisher_name = st.text_input("Publisher Name", key="pub_name_2")
        if st.button("Add Book", key="add_book_btn"):
            insert_book(book_title, publisher_name)

    # Tab 3: Add Library Branch
    with tab3:
        st.header("Insert a New Library Branch")
        branch_name = st.text_input("Branch Name", key="branch_name_3")
        branch_address = st.text_input("Branch Address", key="branch_address_3")
        if st.button("Add Library Branch", key="add_branch_btn"):
            insert_library_branch(branch_name, branch_address)

    # Tab 4: Add Borrower
    with tab4:
        st.header("Insert a New Borrower")
        borrower_name = st.text_input("Borrower Name", key="borrower_name_4")
        borrower_address = st.text_input("Borrower Address", key="borrower_address_4")
        borrower_phone = st.text_input("Borrower Phone", key="borrower_phone_4")
        if st.button("Add Borrower", key="add_borrower_btn"):
            insert_borrower(borrower_name, borrower_address, borrower_phone)

    # Tab 5: Add Book Copies
    with tab5:
        st.header("Insert Book Copies")
        book_id = st.number_input("Book ID", min_value=1, key="book_id_5")
        branch_id = st.number_input("Branch ID", min_value=1, key="branch_id_5")
        no_of_copies = st.number_input("Number of Copies", min_value=1, key="no_of_copies_5")
        if st.button("Add Book Copies", key="add_book_copies_btn"):
            insert_book_copies(book_id, branch_id, no_of_copies)

    # Tab 6: Add Book Authors
    with tab6:
        st.header("Insert a New Book Author")
        book_id = st.number_input("Book ID", min_value=1, key="book_id_6")
        author_name = st.text_input("Author Name", key="author_name_6")
        if st.button("Add Book Author", key="add_book_author_btn"):
            insert_book_authors(book_id, author_name)

    # Tab 7: Issue Book
    with tab7:
        st.header("Issue a Book")
        member = st.text_input("Member Username", key="member_7")
        book_isbn = st.text_input("Book ISBN", key="book_isbn_7")
        price = st.number_input("Price", min_value=0, key="price_7")
        if st.button("Issue Book", key="issue_book_btn"):
            insert_book_issue(member, book_isbn, price)

    # Tab 8: Add Librarian
    with tab8:
        st.header("Insert a New Librarian")
        username = st.text_input("Username", key="username_8")
        password = st.text_input("Password", type="password", key="password_8")
        if st.button("Add Librarian", key="add_librarian_btn"):
            insert_librarian(username, password)

    # Tab 9: Pending Registration
    with tab9:
        st.header("Add a Pending Registration")
        username = st.text_input("Username", key="username_9")
        password = st.text_input("Password", type="password", key="password_9")
        name = st.text_input("Name", key="name_9")
        email = st.text_input("Email", key="email_9")
        balance = st.number_input("Balance", min_value=0, key="balance_9")
        if st.button("Add Pending Registration", key="add_pending_registration_btn"):
            insert_pending_registration(username, password, name, email, balance)

if __name__ == "__main__":
    main()
```
